```python
#Question 4
import PyPDF2

def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        for page_num in range(len(reader.pages)):
            text += reader.pages[page_num].extract_text()
    return text

def main():
    pdf_path1 = '/content/foxnews.com-Embattled Harvard University President Claudine Gay hit with more plagiarism charges.pdf'
    pdf_path2 = '/content/reuters.com-Harvard president will submit dissertation edits after plagiarism allegations.pdf'

    # Extract text from PDFs
    text1 = extract_text_from_pdf(pdf_path1)
    text2 = extract_text_from_pdf(pdf_path2)

    return text1, text2

if __name__ == "__main__":
    text1, text2 = main()
    print("Text from Embattled Harvard University President Claudine Gay hit with more plagiarism charges:")
    print(text1)

    print("\nText from Harvard president will submit dissertation edits after plagiarism allegations:")
    print(text2)
```

The third is incorrect formatting. As we can see, there are missing line breaks and paragraph spacing, which makes the extracted text difficult to read.

```python
#Question 7
from readability import Readability
from readability.exceptions import ReadabilityException

def calculate_readability_scores(text):
    try:
        r = Readability(text)
        fk_score = r.flesch_kincaid()
        gf_score = r.gunning_fog()
        cl_score = r.coleman_liau()
        return fk_score, gf_score, cl_score
    except ReadabilityException as e:
        print("Error:", e)
        return None, None, None

def main():
    # Calculate readability scores for each article
    fk_score1, gf_score1, cl_score1 = calculate_readability_scores(text1)
    fk_score2, gf_score2, cl_score2 = calculate_readability_scores(text2)

    # Print readability scores
    print("Readability scores for article 1:")
    print("Flesch-Kincaid Grade Level:", fk_score1)
    print("Gunning Fog Index:", gf_score1)
    print("Coleman-Liau Index:", cl_score1)
    print("\nReadability scores for article 2:")
    print("Flesch-Kincaid Grade Level:", fk_score2)
    print("Gunning Fog Index:", gf_score2)
    print("Coleman-Liau Index:", cl_score2)

if __name__ == "__main__":
    main()
```

**Flesch-Kincaid Grade Level (FKGL):** This metric estimates the U.S. grade level needed to understand the text. A lower score indicates easier readability.

Article 1: Score of approximately 19.22, corresponding to a grade level of '19'.

Article 2: Score of approximately 18.71, corresponding to a grade level of '19'.

The Flesch-Kincaid Grade Level estimates the U.S. grade level required to understand the text. A grade level of '19' indicates that the text is written at a very high level and may be best understood by individuals with extensive educational backgrounds or specialized knowledge.


**Gunning Fog Index (GFI):** This metric estimates the years of formal education required to understand the text. A lower score indicates easier readability.

Article 1: Score of approximately 21.87, corresponding to a grade level of 'college_graduate'.

Article 2: Score of approximately 19.59, also corresponding to a grade level of 'college_graduate'.

The Gunning Fog Index estimates the number of years of formal education required to understand the text. A grade level of 'college_graduate' suggests that the text is written at a level typically understood by individuals with a college-level education or higher.

**Coleman-Liau Index (CLI):** This metric estimates the U.S. grade level needed to understand the text based on characters per word and words per sentence. A lower score indicates easier readability.

Article 1: Score of approximately 13.41, corresponding to a grade level of '13'.

Article 2: Score of approximately 15.24, corresponding to a grade level of '15'.

The Coleman-Liau Index estimates the U.S. grade level required to understand the text based on characters per word and words per sentence. A grade level of '13' or '15' indicates that the text is written at a level generally understood by individuals in the 13th or 15th grade, respectively.

```python
#Question 8
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def generate_word_cloud(text, title):
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(title)
    plt.axis('off')
    plt.show()

def main():
    # Generate word clouds
    generate_word_cloud(text1, "Word Cloud for Article 1")
    generate_word_cloud(text2, "Word Cloud for Article 2")

if __name__ == "__main__":
    main()
```
